```python
import pandas as pd
import glob

# Biometric
bio_files = glob.glob("data/raw/api_data_aadhar_biometric/*.csv")
bio = pd.concat([pd.read_csv(f) for f in bio_files], ignore_index=True)

# Demographic
demo_files = glob.glob("data/raw/api_data_aadhar_demographic/*.csv")
demo = pd.concat([pd.read_csv(f) for f in demo_files], ignore_index=True)

# Enrolment
enrol_files = glob.glob("data/raw/api_data_aadhar_enrolment/*.csv")
enrol = pd.concat([pd.read_csv(f) for f in enrol_files], ignore_index=True)
```

```python
print(bio.shape)
print(demo.shape)
print(enrol.shape)
```

```python
bio.to_csv("data/processed/biometric_full.csv", index=False)
demo.to_csv("data/processed/demographic_full.csv", index=False)
enrol.to_csv("data/processed/enrolment_full.csv", index=False)
```
